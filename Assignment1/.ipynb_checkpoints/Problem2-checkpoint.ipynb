```python
%pylab inline

from sklearn import datasets

iris = datasets.load_iris()
features = []
for feature in iris.feature_names:
    features.append(iris.data[:,iris.feature_names.index(feature)])

fig, ax = subplots(len(features), len(features), figsize=(10, 10))
lines = [None]*len(iris.target_names)
for x in range(len(features)):
    for y in range(len(features)):
        if x == y:
            ax[y][x].hist(features[x])
        else:
            for target in set(iris.target):
                example_ids = (target == iris.target)
                lines[target] = ax[y][x].scatter(features[x][example_ids], features[y][example_ids],
                                                 label=iris.target_names[target], color='bgr'[target], marker='o')

for i in range(len(features)):
    ax[i][0].set_ylabel(iris.feature_names[i])
    ax[len(features) - 1][i].set_xlabel(iris.feature_names[i])

fig.legend(lines, labels=iris.target_names, loc='upper center', ncol=3)
```
