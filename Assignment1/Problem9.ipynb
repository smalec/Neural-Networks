```python
%pylab inline

from numpy import random

def L(p, k, n):
    return (p**k)*(1 - p)**(n - k)

def bernoulli(n, p):
    trials = random.binomial(1, p, n)
    successes = sum(trials)
    axvline(successes*1.0 / n, color='r', label='sample mean')
    xline = linspace(0.0, 1.0, 1000)
    yline = L(xline, successes, n)
    plot(xline, yline)
    xlabel('p')
    ylabel('L(p; x)')
    legend(loc='upper right')
    title('Likelihood for ' + str(n) + ' Bernoulli trials with p=' + str(p))

bernoulli(1000, 0.4)
```
